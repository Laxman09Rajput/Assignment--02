 8.What is a hash table, and how does it relate to dictionaries in Python ?

      - A hash table is a type of data structure that uses a technique known as hashing to map keys to values.  It makes data retrieval incredibly quick by converting each key into a unique index using a hash function.  Rather than going through each element, the hash table goes straight to the value's storage location.  Hash tables are used in Python to implement dictionaries.  Because of this, dictionary operations such as finding, adding, or removing values are very effective and usually take a constant amount of time (O(1)).  Dictionary hashing requires that its keys be immutable (such as strings, numbers, or tuples) in order to provide consistent hashing

 13. How do sets handle duplicate values in Python ?
     
  - In Python, sets don’t mess around with duplicates. If you try to add the same value more than once, Python just shrugs and keeps only one of them. Simple as that.

             Example   -  {my_set = {1, 2, 3, 3, 4, 4, 5}  
                         print(my_set)  # Output: {1, 2, 3, 4, 5}}



  See what happened? Even though 3 and 4 showed up twice, the set only kept one of each. That’s because sets use a special system under the hood (called a hash table) that automatically ignores duplicates.

 17. Describe the time complexity of accessing elements in a dictionary ?
     
     -  
      Python dictionaries make use of hash tables, which makes it very fast to access an element within one.  It usually takes O(1) time to look up a key, regardless of the dictionary's size.  Sometimes hash collisions cause too many keys to end up in the same place, slowing down the lookup time to O(n).  However, Python's hash table is designed to handle this efficiently, so it's not a common issue.  Because dictionary lookups are almost instantaneous, they are generally far faster than list searches

 19. Why are dictionaries considered unordered, and how does that affect data retrieval ?
  - Unlike lists, which use index-based access, dictionaries require a key to retrieve a value; in Python 3.7 and later, dictionaries maintain insertion order, so items appear in the order they were added when iterated; however, their main focus is still efficient key-based access rather than strict ordering. While dictionaries are great for quick searches, they aren't ideal for index-based retrieval or sorting; if order is important, OrderedDict (for older Python versions) or lists of tuples may be better options. However, modern dictionaries are incredibly efficient for the majority of tasks.
